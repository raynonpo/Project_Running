```python
import pandas as pd
df = pd.read_csv("Compile10KMRun.csv")
df.columns=[i for i in df.columns.str.upper()]
print(df)
df["DATE"] = (
    pd.Timestamp("1900-01-01")+ pd.to_timedelta(df["DATE"] - 693596, unit="D")
).dt.date

# Divide selected columns
divide_by_60 = [
    "TIME",
    "AVG_PACE_MIN_KM",
    "BEST_PACE_MIN_KM",
    "MOVING_TIME",
    "AVG_MOVING_PACE_MIN_KM"
]

df[divide_by_60] = df[divide_by_60] / 60
df["CUMULATIVE_TIME"] = df["CUMULATIVE_TIME"] / 50

# Select columns in the desired order
stage_df = df[
    [
        "Date",
        "MONTH",
        "DAY",
        "LAPS",
        "TIME",
        "CONDITION_CATEGORY",
        "CUMULATIVE_TIME",
        "AVG_PACE_MIN_KM",
        "AVG_HR_BPM",
        "MAX_HR_BPM",
        "TOTAL_ASCENT_M",
        "TOTAL_DESCENT_M",
        "AVG_RUN_CADENCE_SPM",
        "AVG_STRIDE_LENGTH_M",
        "CALORIES_C",
        "BEST_PACE_MIN_KM",
        "MAX_RUN_CADENCE_SPM",
        "MOVING_TIME",
        "AVG_MOVING_PACE_MIN_KM",
    ]
]

stage_df.to_csv("STAGE_RUNPROJECT.csv", index=False)
```
